```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_errs(c, s):
    return np.loadtxt('binary_lds_c' + str(c) + '_s' + str(s) + '_errs.txt')
```

```python
def load_elbos(c, s):
    return np.loadtxt('binary_lds_c' + str(c) + '_s' + str(s) + '_elbo.txt')
```

```python
def load_all_errs(c):
    return np.stack([load_errs(c, s) for s in range(5)], axis=1)
```

```python
def load_all_elbos(c):
    return np.stack([load_elbos(c, s) for s in range(5)], axis=1)
```

```python
errs_c1 = load_all_errs(1)
errs_c2 = load_all_errs(2)
errs_c3 = load_all_errs(3)
```

```python
elbos_c1 = load_all_elbos(1)
elbos_c2 = load_all_elbos(2)
elbos_c3 = load_all_elbos(3)
```

```python
plt.semilogy(np.arange(0, 2000, 10), np.mean(errs_c1, axis=-1), color='C3')
plt.semilogy(np.arange(0, 2000, 10), np.mean(errs_c2, axis=-1), color='C2')
#plt.semilogy(np.arange(0, 2000, 10), np.mean(errs_c3, axis=-1), color='C6')
plt.fill_between(range(0, 2000, 10), np.min(errs_c1, axis=-1), np.max(errs_c1, axis=-1), alpha=0.5, color='C3')
plt.fill_between(range(0, 2000, 10), np.min(errs_c2, axis=-1), np.max(errs_c2, axis=-1), alpha=0.5, color='C2')
#plt.fill_between(range(0, 2000, 10), np.min(errs_c3, axis=-1), np.max(errs_c3, axis=-1), alpha=0.5, color='C6')
plt.ylim(3.6e-1, 4e-1)
```

```python
lines1 = plt.semilogy(np.arange(0, 2000, 10), errs_c1, color='C3', label='SVAE')
lines2 = plt.semilogy(np.arange(0, 2000, 10), errs_c2, color='C2', label='EPSVAE')
plt.yticks([3.5e-1, 3.6e-1, 3.7e-1, 3.8e-1, 3.9e-1])

plt.legend([lines1[0], lines2[0]], ['SVAE', 'EP-SVAE (M=2)'])
#plt.semilogy(np.arange(0, 2000, 10), np.mean(errs_c3, axis=-1), color='C6')
plt.ylim(3.6e-1, 4e-1)
```

```python
plt.semilogy(np.arange(0, 2000, 10), np.mean(elbos_c1, axis=-1), color='C3')
plt.semilogy(np.arange(0, 2000, 10), np.mean(elbos_c2, axis=-1), color='C2')
plt.labels(['SVAE', 'EPSVAE MoG 2'])
#plt.semilogy(np.arange(0, 2000, 10), np.mean(elbos_c3, axis=-1), color='C6')
#plt.fill_between(range(0, 2000, 10), np.min(elbos_c1, axis=-1), np.max(elbos_c1, axis=-1), alpha=0.5, color='C3')
#plt.fill_between(range(0, 2000, 10), np.min(elbos_c2, axis=-1), np.max(elbos_c2, axis=-1), alpha=0.5, color='C2')
#plt.fill_between(range(0, 2000, 10), np.min(elbos_c3, axis=-1), np.max(elbos_c3, axis=-1), alpha=0.5, color='C6')
plt.ylim(4.5e+1, 6e+1)
```
